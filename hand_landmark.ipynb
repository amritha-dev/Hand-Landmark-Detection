```python
import cv2 as cv
import mediapipe.python.solutions.hands as mpHands
import mediapipe.python.solutions.drawing_utils as drawing
```

```python
hands=mpHands.Hands(
   static_image_mode=False,max_num_hands=2,min_detection_confidence=0.5
)
```

```python
source=cv.VideoCapture(0)
```

```python
while True:
    success, frame=source.read()
    if not success:
        print('Camera not detected...!')
        
    frame=cv.flip(frame,1) 
    frame_RGB=cv.cvtColor(frame, cv.COLOR_BGR2RGB)
    hands_detect=hands.process(frame_RGB)
        
    cv.imshow("Hand Landmark",frame)
    if hands_detect.multi_hand_landmarks:
        for landmark in hands_detect.multi_hand_landmarks:
            print(landmark)
            drawing.draw_landmarks(image=frame,landmark_list=landmark,connections=mpHands.HAND_CONNECTIONS)
    cv.imshow("Hand Landmark",frame)       
    
    if cv.waitKey(1) == ord('q'):            
        break
cv.destroyAllWindows()
source.release()  
```
